# Lab Task 01

Reshape a 1D NumPy array into a 2D array with dimensions (3, 4).

Concatenate two NumPy arrays horizontally and vertically.

```python
import numpy as np

original_array = np.random.randint(10, size=12)
print("original array :")
print(original_array)
reshaped_array = original_array.reshape(3, 4)
print("Reshaped array:")
print(reshaped_array)

# Concatenate two NumPy arrays horizontally
array1 = np.random.randint(10, size=(2, 2))
array2 = np.random.randint(10, size=(2, 2))
print("\narray1:")
print(array1)
print("array1:")
print(array1)
horizontal_concat = np.concatenate((array1, array2), axis=1)
print("\nHorizontally concatenated array:")
print(horizontal_concat)

# Concatenate two NumPy arrays vertically
array1 = np.random.randint(10, size=(2, 2))
array2 = np.random.randint(10, size=(2, 2))
print("\narray1:")
print(array1)
print("array1:")
print(array1)
vertical_concat = np.concatenate((array1, array2), axis=0)
print("\nVertically concatenated array:")
print(vertical_concat)
```

## Email Classification
Develope a tool which will classify an email.

### Tasks:
* Create a function classify_emails that takes a list of emails as an argument.
* Inside the function, categorize emails into "Important," "Promotions," and "Spam" based on keywords and content analysis.
* Loop through the emails and identify keywords to determine their category.
* If an email contains keywords indicating importance, assign it as "Important."
* If an email contains keywords related to promotions or deals, assign it as "Promotions."
* If an email is suspected to be spam based on certain keywords or patterns, assign it as "Spam."
* Return three lists of emails for each category.
* Print the lists.

```python
import re

def classify_emails(emails):
  important_emails = []
  promotional_emails = []
  spam_emails = []

  for email in emails:
    if re.search(r"(important|urgent|priority)", email, re.IGNORECASE):
      important_emails.append(email)
    elif re.search(r"(sale|discount|offer|promotion)", email, re.IGNORECASE):
      promotional_emails.append(email)
    elif re.search(r"(viagra|casino|free money|click here)", email, re.IGNORECASE):
      spam_emails.append(email)
  return important_emails, promotional_emails, spam_emails

emails = ["Important: Meeting Agenda", "50% Off Sale!", "Get Rich Quick Scheme", "Urgent: Project Update", "Free iPhone!"]
important_emails, promotional_emails, spam_emails = classify_emails(emails)

print("Important emails:")
for email in important_emails:
  print(email)

print("\nPromotional emails:")
for email in promotional_emails:
  print(email)

print("\nSpam emails:")
for email in spam_emails:
  print(email)
```

## English-Spanish Dictionary
You are creating a basic English-Spanish dictionary that allows users to look up English words and find their Spanish translations.


### Tasks
* Create a dictionary where English words are keys and their Spanish translations are values.
* Ask the user to input an English word.
* Check if the word exists in the dictionary.
* If the word exists, print its Spanish translation.
* If the word does not exist, inform the user that the translation is not available.
* Implement a loop that allows the user to keep looking up words until they choose to exit.


```python
#code here
# Create a dictionary of English words and their Spanish translations
english_spanish_dictionary = {
    "hello": "hola",
    "goodbye": "adiós",
    "thank you": "gracias",
    "you're welcome": "de nada",
    "yes": "sí",
    "no": "no",
    "dog": "perro",
    "cat": "gato",
    "house": "casa",
    "car": "coche"
}

# Get the user's input
while True:
    english_word = input("Enter an English word (or type 'exit' to quit): ")

    # Check if the word exists in the dictionary
    if english_word in english_spanish_dictionary:
        # Print the Spanish translation
        spanish_translation = english_spanish_dictionary[english_word]
        print(f"The Spanish translation of '{english_word}' is '{spanish_translation}'.")
    elif english_word == 'exit':
        break
    else:
        # Inform the user that the translation is not available
        print(f"Sorry, the translation for '{english_word}' is not available in the dictionary.")
```

```python
import pandas as pd
data = pd.read_csv("/content/vgsalesGlobale.csv")
data
```

```python
total_global_sales = data['Global_Sales'].sum()

highest_selling_game = data[data['Global_Sales'] == data['Global_Sales'].max()]

average_sales_by_region =data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].mean(axis=0)

top_5_genres = data.groupby('Genre').sum()['Global_Sales'].nlargest(5)

total_sales_by_genre_region = data.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
most_popular_genre_na = total_sales_by_genre_region.sort_values(by='NA_Sales', ascending=False).index[0]

last = data.tail(10)

top_3_platforms = data.groupby('Platform').sum()['Global_Sales'].nlargest(3)

market_share_by_genre = (data.groupby('Genre').sum()['Global_Sales'] / total_global_sales) * 100

correlation_na_eu = data[['NA_Sales', 'EU_Sales']].corr().iloc[0, 1]


print("Total global sales:", total_global_sales)
print("Game with highest global sales:")
print(highest_selling_game)
print("Average sales by region:")
print(average_sales_by_region)
print("Top 5 most popular game genres:")
print(top_5_genres)
print("Most popular genre in North America:", most_popular_genre_na)
print("Last 10 rows of the dataset:")
print(last)
print("Top 3 gaming platforms with highest global sales:")
print(top_3_platforms)
print("Market share by genre:")
print(market_share_by_genre)
print("Correlation coefficient between NA_Sales and EU_Sales: \n", correlation_na_eu)
```
